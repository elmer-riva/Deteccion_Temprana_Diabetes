## Detección de outliers

Identificare posibles outliers en las variables numéricas, posiblemente utilizando métodos estadísticos (como box plots o el método del rango intercuartílico).


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Seleccionar las columnas numéricas
columnas_numericas = df.select_dtypes(include=['float64', 'int64']).columns

# Excluir las columnas binarias que representan categorías (aunque sean numéricas)
columnas_numericas = columnas_numericas.drop(['hypertension', 'heart_disease', 'diabetes'])

# Crear box plots para cada columna numérica seleccionada
for columna in columnas_numericas:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=df[columna])
    plt.title(f'Box Plot de {columna}')
    plt.xlabel(columna)
    plt.show()
```

### Observaciones sobre outliers en variables numéricas

Después de visualizar las distribuciones de las variables numéricas ('age', 'bmi', 'HbA1c_level', 'blood_glucose_level') utilizando box plots, observé lo siguiente:

*   **age:** El box plot para la edad no muestra outliers. La distribución parece estar dentro del rango esperado para una población diversa.
*   **bmi:** El box plot para el IMC (bmi) muestra una cantidad significativa de puntos individuales por encima del bigote superior del box plot, lo que indica la presencia de numerosos outliers con valores de IMC elevados. Esto concuerda con mi observación previa de un valor máximo de 95.69 en las estadísticas descriptivas. Estos outliers representan individuos con obesidad extrema.
*   **HbA1c_level:** El box plot para el nivel de HbA1c no muestra outliers significativos. La mayoría de los puntos se encuentran dentro o muy cerca del rango intercuartílico y sus bigotes.
*   **blood_glucose_level:** El box plot para el nivel de glucosa en sangre muestra un número considerable de puntos individuales por encima del bigote superior, indicando la presencia de outliers con niveles de glucosa en sangre elevados. Esto es esperable en un dataset relacionado con la diabetes.

En resumen, identifiqué outliers significativos en las variables `bmi` y `blood_glucose_level`. La presencia de estos outliers en `bmi` (obesidad extrema) y `blood_glucose_level` (glucosa alta) es relevante para el análisis predictivo de la diabetes y podría requerir una consideración especial durante el preprocesamiento de los datos, dependiendo del modelo a utilizar.

**Posibles estrategias simples para manejar los outliers:**

Dado que los outliers en `bmi` y `blood_glucose_level` son valores reales y relevantes para la predicción de diabetes, eliminarlos por completo podría resultar en la pérdida de información importante.
    *   Podría **limitar** los valores extremos: por ejemplo, si un valor de `bmi` es extremadamente alto (como 95.69), podría decidir tratarlo como un valor un poco menos extremo pero aún alto (como 60 o 70), en lugar de eliminar la fila completa. Esto se llama "truncamiento".
    *   Si solo son muy pocos valores **extremadamente** fuera de lo común y no parecen correctos (aunque en este caso sí parecen biológicamente posibles), podría considerar eliminarlos, pero esto debe hacerse con mucha precaución para no perder información valiosa.
    *   Otra opción sencilla es **reemplazar los valores atípicos con la media o la mediana** de la columna. Por ejemplo, si un valor de `bmi` es un outlier, podría cambiarlo por el valor promedio de `bmi` en todo el dataset. Sin embargo, debo tener en cuenta que esto puede distorsionar la distribución original de los datos y reducir la variabilidad, especialmente si hay muchos outliers. La mediana suele ser menos sensible a los valores extremos que la media.

## Visualizaciones principales


Primera visualizacion, un hiastograma para distribucion de 'age'



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograma para la distribución de 'age'
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', kde=True, bins=30)
plt.title('Distribución de la Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

Creer la segunda visualización, un diagrama de caja para 'HbA1c_level' segmentado por 'diabetes', y la tercera visualización, un diagrama de conteo para 'smoking_history' segmentado por 'diabetes'.


```python
# Box plot para 'HbA1c_level' segmentado por 'diabetes'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='diabetes', y='HbA1c_level')
plt.title('Nivel de HbA1c por Estado de Diabetes')
plt.xlabel('Diabetes (0: No, 1: Sí)')
plt.ylabel('Nivel de HbA1c')
plt.xticks([0, 1], ['No Diabetes', 'Diabetes'])
plt.show()

# Countplot para 'smoking_history' segmentado por 'diabetes'
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='smoking_history', hue='diabetes', palette='viridis')
plt.title('Conteo por Historial de Tabaquismo y Estado de Diabetes')
plt.xlabel('Historial de Tabaquismo')
plt.ylabel('Conteo')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Diabetes', labels=['No', 'Sí'])
plt.tight_layout()
plt.show()
```
